```python
# higher order function-- takes one or more fumctions as arguments
--filter
--map
--reduce
--enumerate
```

```python
# filter(function,variable)
num=[7,6,8,4,5,6]
even=list(filter(lambda x:x%2==0,num))
print(even)
```

```python
# map pblm -- i want only bonus of each 4 to be added.
salary=[100,200,300,400]
new_salary=list(map(lambda x: x+4,salary))
print(new_salary)
```

```python
# recursive function--- a function that call by itself
-- sum of n numbers
--factorial number
```

```python
def add(n):
    if n==1:# base condition
        return 1
    else:
        return n +add(n-1)
add(3)
```

```python
add(3) n=3 3+add(2)  3+3=6   -- substitute the value
add(2) n=2 2+add(1)  2+1=3    
add(1) n=1 1
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(4))
```
