```python
# 4.write a function to sort the list--
num1=[10,14,8,32,9,4,6]
def sort_list(num1):
    for a in range(len(num1)):
        for b in range(a + 1, len(num1)):
            if num1[a] > num1[b]:
                num1[a], num1[b] = num1[b], num1[a]
    return num1
print(sort_list(num1))
```

```python
#5. write a function to reverse the list-
num1 = [10, 14, 8, 32, 9, 4, 6]

def reverse_list(num1):
    rev = []
    for i in range(len(num1)-1, -1, -1):
        rev.append(num1[i])
    return rev

print(reverse_list(num1))
```

```python
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        print("addition:", self.a + self.b)

    def subtract(self):
        print("subtraction:", self.a - self.b)

    def multiply(self):
        print("multiplication:", self.a * self.b)

    def divide(self):
        if self.b == 0:
            print("not divide by zero")
        else:
            print("division:", self.a / self.b)




calc = Calculator(90, 5)
calc.add()
calc.subtract()
calc.multiply()
calc.divide()







```

```python
#shape class with subclasses (circle, triangle, square)

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print("circle area:", 3.14 * self.radius * self.radius)

    def perimeter(self):
        print("circle perimeter:", 2 * 3.14 * self.radius)

class Square:
    def __init__(self, side):
        self.side = side

    def area(self):
        print("square area:", self.side * self.side)

    def perimeter(self):
        print("square perimeter:", 4 * self.side)

class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        p = self.a + self.b + self.c
        print("triangle perimeter:", p)

    def area(self):
        s = (self.a + self.b + self.c) / 2  
        area = (s * (s - self.a) * (s - self.b) * (s - self.c)) ** 0.5
        print("triangle area:", area)


circle = Circle(5)
circle.area()
circle.perimeter()

square = Square(4)
square.area()
square.perimeter()

triangle = Triangle(3, 4, 5)
triangle.area()
triangle.perimeter()
```

```python
#Write a Python program to create a class representing a shopping cart.
#Include methods for adding and removing items, and calculating the total price.

items = {}

def add_item(name, price):
    items[name] = price
    print(name, "added")

def remove_item(name):
    if name in items:
        del items[name]
        print(name, "removed")
    else:
        print(name, "not found")

def total_price():
    total = 0
    for price in items.values():
        total = total+price
    print("total price:", total)


add_item("butter", 30)
add_item("milk", 50)
total_price()
remove_item("apple")
total_price()
```

```python
import pandas as pd
df1=pd.read_excel("C:/Users/Sathyanarayanan/Downloads/Sales Analysis Report (1) (1) (1).xlsx")
df1.head()
```

```python
import pandas as pd
import scipy.stats as stats

df1["Channel"] = df1["Channel"].astype("category")

Wholesale= df1[df1["Channel"] =="Wholesale"]["Order Quantity"]
Distributor= df1[df1["Channel"] =="Distributor"]["Order Quantity"]
Export=df1[df1["Channel"] =="Export"]["Order Quantity"]

anova_result = stats.f_oneway(Wholesale, Distributor, Export)

print(f"F-statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")

if anova_result.pvalue < 0.05:
    print("Their relation is strong")
else:
    print("Their relation is  not strong")
```

```python
# common keys 
dict_1={"a":10,"b":20,"c":30}
dict_2={"a":13,"f":23,"j":45}

common_keys = []

for k in dict_1:
    if k in dict_2:
        common_keys.append(k)

print(common_keys)
```

```python
#sort using keys
dict_2={1001:"a",1007:"b",1002:"c"}

keys = list(dict_2.keys())

for i in range(len(keys)):
    for j in range(i+1, len(keys)):
        if keys[i] > keys[j]:    
            keys[i], keys[j] = keys[j], keys[i]

sorted_dict = {}
for k in keys:
    sorted_dict[k] = dict_2[k]

print(sorted_dict)
```

```python
#remove duplicates using dictionary
numbers=[1,1,1,2,2,3]

unique_dict = {}
for n in numbers:
    unique_dict[n] = True    

unique_list = list(unique_dict.keys())
print(unique_list)
```

```python
# remove duplicate values from the dictionary

dict_1 = {"a":13, "b":20, "c":30, "f":20}

count = {}
for v in dict_1.values():
    count[v] = count.get(v, 0) + 1
    
unique_keys = []
for k, v in dict_1.items():
    if count[v] == 1:
        unique_keys.append(k)

print(unique_keys)
```
